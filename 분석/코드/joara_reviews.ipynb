[Selenium 사용] 더보기 버튼이 없을 때까지 클릭 후 댓글 작성 날짜와 내용, 총 댓글 수 추출

## 디비에서 리디북스 URL 가져와 대입하여 댓글 수집

# 코드 제작자 : 한승주

```python
import urllib.request
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
from urllib import parse
import time

url = "http://www.joara.com/premium_new/book_intro.html?book_code=1453104&list_type=&book_dtype=comment_premium"
driver = webdriver.Chrome("C:/project/chromedriver.exe")
driver.get(url)
time.sleep(1)

# 더보기 버튼 클릭하기
while True:
    try:
        more_btn = driver.find_element_by_css_selector('#commentMoreBtn > a')
        more_btn.click()
        time.sleep(1)
        
    except:
        break
        
# 댓글 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
comment_list = soup.select('#comment_list')

comments = comment_list[0].find_all(name="dd")

review_data = []
total = 0
for comment in comments:
    date = comment.find('span',class_='date').text[0:10]
    print(date)

    review=comment.find('p',class_='comment').text
    print(review)
    review_data.append([date, review])
    total += 1
    
print("총 댓글 수: ", total)
driver.close()

# total : 총 댓글수 | review_data [type: list]
```

```python
# 수집한 정보 DataFrame으로 확인하기
import pandas as pd

df=pd.DataFrame(review_data)
df
```
