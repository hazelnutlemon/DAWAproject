```python
import bs4
import requests
from urllib import parse
import re
import pandas as pd

def tistory_crawling(search_name):
    session = requests.Session()
    header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36"}

    tmp = search_name + "리뷰" #혹은 후기
    base_url = "https://search.daum.net/search?w=blog&nil_search=btn&DA=STC&enc=utf8&q="+parse.quote(tmp) + "&f=section&SA=tistory&page={}"
    n=1
    pre_page=None
    
    # initialize
    info_lists = []
    
    while True:
        url=base_url.format(n)
        r=requests.get(url)
        c=r.content
        html = bs4.BeautifulSoup(c, "html.parser")

        main=html.find('div',{'class','type_fulltext wid_n'})

        count=html.find('span',{'class','txt_info'})
        if(count==None):
            print("프로그램 종료합니다")
            break
        page=count.text.split(" /")[0]
        if(pre_page==page):
            print("프로그램 종료합니다")
            break
        pre_page=page
        n=n+1
        main=html.find('div',{'class','type_fulltext wid_n'})
        coll=main.find('div',{'class','coll_cont'})
        ul=coll.find('ul')
        li_s=ul.findAll('li')

        for li in li_s:
            href=li.find('a').attrs['href'].replace(".com/",".com/m/")
            if 'tistory' not in href:
                continue
            r=session.get(href,headers=header)
            html = bs4.BeautifulSoup(r.content, "html.parser")

            title=html.find('h2',{'class','tit_blogview'}).text
            if search_name.replace(" ","") not in title.replace(" ",""):
                continue
            print("제목 : " + title)

            author=html.find('span',{'class','txt_by'}).text
            print("작성자 : " + author)

            date=html.find('time',{'class','txt_date'}).text
            print("작성일시 : " + date)

            post_area=html.find('div',{'class','blogview_content'})

            texts=post_area.findAll('p')
            post=''
            print("\n----------------------------------------\n" + "게시글\n" + "----------------------------------------\n")
            for text in texts:
                post=post+text.text
            print(post)
            print("\n----------------------------------------\n")
            post=re.sub(r'[^ ㄱ-ㅎ|ㅏ-ㅣ|가-힣a-zA-Z0-9\:\/\-\+\*\~\!\`\'\"\!\@\#\$\%\^\&\*\(\.\,\<\/\?\:\;\[\]\{\}\\\|]', '', post)
            info_list = [title, author, date, post]
    
            info_lists.append(info_list)

    
    df = pd.DataFrame(info_lists, columns = ["title", "author", "date", "post"])

    # csv 파일 만들기
    df.to_csv("{}.csv".format(search_name), index=False, encoding='cp949')
    print("저장 완료.\n")
```

```python
tistory_crawling('닥터최태수')
```
