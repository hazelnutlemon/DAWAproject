# 리디북스 댓글
[Selenium 사용] 로그인 후 작품 페이지로 들어가 더보기 버튼이 나오지 않을 때까지 클릭.
댓글작성 날짜, 댓글 내용, 총 댓글 개수를 추출

```python
import urllib.request
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
from urllib import parse
import time

driver = webdriver.Chrome("C:/project/chromedriver.exe")

# 로그인 계정 정보
ridi_id = 'shung2rhea'
ridi_pwd = 'rhea155204'

# 로그인
driver.get('https://ridibooks.com/account/login')
time.sleep(3)
# ID 입력
ridi_login = driver.find_element_by_id("login_id")
ridi_login.clear()
ridi_login.send_keys(ridi_id)
# 비밀번호 입력
ridi_login = driver.find_element_by_id("login_pw")
ridi_login.clear()
ridi_login.send_keys(ridi_pwd)
# 로그인 버튼 클릭
driver.find_element_by_xpath("/html/body/div[1]/div/div/section/section/form/button").click()

# 댓글 크롤링
url = "https://ridibooks.com/books/3214012687"
time.sleep(3)
driver.get(url)


# 전체 댓글 버튼 클릭
driver.find_element_by_xpath("/html/body/div[3]/div[1]/div/div/div[1]/div[1]/section/article[6]/div[3]/div[1]/ul[1]/li[2]/a").click()
time.sleep(3)

# 더보기 버튼 클릭
while True:
    try:
        more_btn = driver.find_element_by_css_selector('#review_list_section > div.review_list_wrapper.js_review_list_wrapper.active > div.more_review_button_wrapper.js_more_review_button_wrapper > button')
        more_btn.click()
        time.sleep(1)
        
    except:
        break
        
# 댓글 가져오기
#date = driver.find_elements_by_css_selector('#review_list_section > div.review_list_wrapper.js_review_list_wrapper.active > ul > li > div.list_left.js_review_info_wrapper > div > ul > li.review_date')
#comments = driver.find_elements_by_css_selector('#review_list_section > div.review_list_wrapper.js_review_list_wrapper.active > ul > li > div.list_right.js_review_wrapper > p')
total = 0

source = driver.page_source
soup=BeautifulSoup(source, 'html.parser')

reviews=soup.select("#review_list_section > div.review_list_wrapper.js_review_list_wrapper.active > ul > li")

for review in reviews:
    date = review.find('li',class_='review_date').text
    date=date.replace(" ","")
    date=date.replace(".","-")[0:10]
    print(date)
    
    if (review.find('span',class_='hidden')==None):
        comment=review.find('p',class_='review_content js_review_content').text
    else:
        comment=review.find('span',class_='hidden').text
    print(comment)
    total += 1
    
print("총 댓글 수: ", total)
driver.close()
```

# 조아라 댓글
[Selenium 사용] 더보기 버튼이 없을 때까지 클릭 후
댓글 작성 날짜와 내용, 총 댓글 수 추출

```python
import urllib.request
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
from urllib import parse
import time

url = "http://www.joara.com/premium_new/book_intro.html?book_code=1453104&list_type=&book_dtype=comment_premium"
driver = webdriver.Chrome("C:/project/chromedriver.exe")
driver.get(url)
time.sleep(1)

# 더보기 버튼 클릭하기
while True:
    try:
        more_btn = driver.find_element_by_css_selector('#commentMoreBtn > a')
        more_btn.click()
        time.sleep(1)
        
    except:
        break
        
# 댓글 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
comment_list = soup.select('#comment_list')

comments = comment_list[0].find_all(name="dd")

total = 0
for comment in comments:
    date = comment.find('span',class_='date').text[0:10]
    print(date)

    review=comment.find('p',class_='comment').text
    print(review)
    total += 1
    
print("총 댓글 수: ", total)
driver.close()
```
